## 전동휠체어 데이터 불러오기

```python
import pandas as pd
Wheelchair_data=pd.read_csv("./data/National_Electric_Wheelchair.csv")
Wheelchair_data.head()
```

## 지도 표현하기
<hr>

### 지도 객체 생성

```python
import folium
map_osm = folium.Map(location=[37.566647,127.978426], zoom_start=8)
map_osm
```

```python
# 위도 경도 Nan값 처리
Wheelchair_data_clean=Wheelchair_data.dropna(axis=0)

for i in Wheelchair_data_clean.index:
    lat = Wheelchair_data_clean.loc[i,'위도']
    lng = Wheelchair_data_clean.loc[i,'경도']
    
    #추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng])
    marker.add_to(map_osm)
    
map_osm
```
